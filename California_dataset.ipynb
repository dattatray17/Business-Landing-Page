```python
#Question 6
#function which calculates median of data points in housing dataset
data = df['total_rooms']

def calculate_median(data):
    # Sorting the data from total_rooms column
    sorted_data = sorted(data)

    # Check if the dataset has an odd or even number of values
    n = len(sorted_data)
    if n % 2 == 1:
        # If odd, the median is the middle data point.
        median = sorted_data[n // 2]
    else:
        # If even, the median is the average of the middle two data points.
        middle1 = sorted_data[(n // 2) - 1]
        middle2 = sorted_data[n // 2]
        median = (middle1 + middle2) / 2.0
    return median

result = calculate_median(data)
print(f"Median: {result}")
```
